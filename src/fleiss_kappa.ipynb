```python
import pandas as pd

df_uyen = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Uyên')
df_tuan = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Tuấn')
df_anh = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Ánh')
df_tam = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Tâm')
df_thu = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Thư')
df_thuy = pd.read_excel(path.join(val_path,'val_annotated_data_1500.xlsx'), sheet_name='Thuy')
```

```python
data = {
    'df_uyen': df_uyen['star'],
    'df_tuan': df_tuan['star'],
    'df_anh': df_anh['star'],
    'df_tam': df_tam['star'],
    'df_thu': df_thu['star'],
    'df_thuy': df_thuy['star'],
    'clean_cmt': df_thuy['clean_cmt']
}

df = pd.DataFrame(data = data)

df
```

```python
occurrences = {col: df[col].value_counts() for col in df[df.columns.difference(['clean_cmt'])]}

# Convert the occurrences dictionary to a DataFrame
occurrences_df = pd.DataFrame(occurrences)

# Display the DataFrame
occurrences_df
```

```python
from statsmodels.stats.inter_rater import aggregate_raters, fleiss_kappa

fleiss_matrix, categories = aggregate_raters(df.drop(['clean_cmt'], axis = 1))
kappa = fleiss_kappa(fleiss_matrix)
kappa
```
